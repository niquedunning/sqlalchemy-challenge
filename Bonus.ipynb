```python
import datetime as dt
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
import pprint
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
measure = Base.classes.measurement
station = Base.classes.station
```

```python
session = Session(engine)
```

```python
#June and December Temps analysis
june = '06'

tempjune = session.query(func.round(func.avg(measure.tobs))).\
    filter(func.strftime("%m", measure.date) == june).all()
tempjune
```

```python
dec = '12'

tempdec = session.query(func.round(func.avg(measure.tobs))).\
    filter(func.strftime("%m", measure.date) == dec).all()
tempdec
```

```python
import scipy.stats as stats

june = '06'

listjune = session.query(measure.tobs).\
    filter(func.strftime("%m", measure.date) == june).all()
listjune

dec = '12'

listdec = session.query(measure.tobs).\
    filter(func.strftime("%m", measure.date) == dec).all()
listdec


ttest = t_stat, p_val = stats.ttest_ind(listjune, listdec, equal_var=False)
ttest
```

```python
#The P-Value is significantly higher than .05. Therefore there is high correlation between the temps taken in June
#and the temps taken in December. 
#I used an unpaired t-test because the sample sizes were different.
```
